```python
import pandas as pd
df = pd.read_csv('features.csv')
df
```

```python
for i in df.columns:
    if i in df['Call']:
        df[i].fillna(df[i].median(), inplace=True)
    else: 
        df[i].fillna(df[i].mode()[0], inplace=True)
        
print(df.isnull().sum(), '\n')
```

```python
df['Call'].unique()
```

```python
df = pd.get_dummies(df, prefix= ['Sex','Age','Call'], columns=['Sex','Age','Call'])
df
```

```python
from sklearn.preprocessing import StandardScaler
features = df.drop('Species',axis=1).columns.values
x=df.loc[:,features].values
y=df.loc[:,['Species']].values
x=StandardScaler().fit_transform(x)
```

```python
SD_df = pd.DataFrame(x,columns = features)
SD_df
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)

principalDf = pd.DataFrame(data = principalComponents,columns=['principal component 1','principal component 2'])
```

```python
df['Species'].unique()
```

```python
from matplotlib import pyplot as plt

fig,ax = plt.subplots()
print(x.shape[0])
for i in range(x.shape[0]):
    if y[i] == 'A. jubatus':
        ax.scatter(principalDf['principal component 1'][i],principalDf['principal component 2'][i],c='r')
    elif y[i] == 'Domestic Cat':
        ax.scatter(principalDf['principal component 1'][i],principalDf['principal component 2'][i],c='g')
    elif y[i] == 'Caracal Carcal':
        ax.scatter(principalDf['principal component 1'][i],principalDf['principal component 2'][i],c='y')
    elif y[i] == 'L. lynx':
        ax.scatter(principalDf['principal component 1'][i],principalDf['principal component 2'][i],c='black')
    elif y[i] == 'Bobcat':
        ax.scatter(principalDf['principal component 1'][i],principalDf['principal component 2'][i],c='purple')    
    else:
        ax.scatter(principalDf['principal component 1'][i],principalDf['principal component 2'][i],c='b')  

ax.set_xlabel('Principal Component 1',fontsize = 15)
ax.set_ylabel('Principal Component 2',fontsize = 15)
ax.set_title('2 Component PCA',fontsize = 20)

targets = ['A. jubatus', 'Domestic Cat', 'Caracal Carcal', 'L. lynx','Bobcat', 'Ocelot'] 
ax.legend(targets)
ax.grid(True)

leg = ax.get_legend()
leg.legendHandles[0].set_color('r')
leg.legendHandles[1].set_color('g')
leg.legendHandles[2].set_color('y')
leg.legendHandles[3].set_color('black')
leg.legendHandles[4].set_color('purple')
leg.legendHandles[5].set_color('b')
plt.show()
```
